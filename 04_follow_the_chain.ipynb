## 嘗試

點了那個連結 ```linkedlist.php?nothing=12345```，頁面寫著：
> and the next nothing is 44827

感覺應該是要用這個數字替換掉網址上的 ```nothing``` 參數，持續的取得下一個資訊，照提示說的應該 400 次就夠了

````python
from urllib import request

baseurl = "http://www.pythonchallenge.com/pc/def/linkedlist.php?nothing="
nothing = "12345"

for i in range(400):
    content = request.urlopen(baseurl + nothing).read().decode()
    print(content)
    nothing = content.split()[-1]
````

在第 84 ~ 88 次執行時會出現
```
and the next nothing is 3875
and the next nothing is 16044
Yes. Divide by two and keep going.
and the next nothing is 55274
and the next nothing is 93397
```

看起來在遇到 ```Yes. Divide by two and keep going.``` 這訊息時應該要把 ```nothing``` 設為 ```16044//2``` 再跑才對
