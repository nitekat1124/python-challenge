## 嘗試

我把 channel.zip 及解壓縮後的 channel 目錄都先丟進 data 目錄

````python
nothing = "90052"
while 1:
    with open(f"./data/channel/{nothing}.txt") as f:
        content = f.read()
    print(content)
    nothing = content.split()[-1]
    if not nothing.isdigit():
        break
````

zip comment ??

````python
from zipfile import ZipFile

zip = ZipFile("./data/channel.zip")
nothing = "90052"

while 1:
    zip_info = zip.getinfo(f"{nothing}.txt")
    print(zip_info.comment.decode(), end="")
    content = zip.read(zip_info).decode()
    # print(content)
    nothing = content.split()[-1]
    if not nothing.isdigit():
        break
````

因此可以得知下一關的網址就是 ```http://www.pythonchallenge.com/pc/def/hockey.html```
