## 嘗試

既然標題是 odd even，那麼我們就將圖拆成兩份來看看好了

````python
from PIL import Image, ImageEnhance
import urllib.request

baseurl = "http://www.pythonchallenge.com/pc/return/"

pwmgr = urllib.request.HTTPPasswordMgrWithDefaultRealm()
pwmgr.add_password(None, baseurl, "huge", "file")
auth_handler = urllib.request.HTTPBasicAuthHandler(pwmgr)
opener = urllib.request.build_opener(auth_handler)
urllib.request.install_opener(opener)

url = "http://www.pythonchallenge.com/pc/return/cave.jpg"

data = urllib.request.urlopen(url)
im = Image.open(data)

im_odd = Image.new("RGB", (im.width, im.height), "black")
im_even = Image.new("RGB", (im.width, im.height), "black")

for y in range(im.height):
    for x in range(im.width):
        if (x + y) % 2 == 0:
            im_even.putpixel((x, y), im.getpixel((x, y)))
        else:
            im_odd.putpixel((x, y), im.getpixel((x, y)))

display(im_odd)
display(im_even)
````

第二張圖好像有什麼東西，調亮一點試試看

````python
display(ImageEnhance.Brightness(im_even).enhance(4))
````
