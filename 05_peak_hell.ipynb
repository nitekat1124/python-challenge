# 嘗試

讀取 ```banner.p``` 的內容後用 pickle 反序列化

````python
from urllib.request import urlopen
import pickle

u = "http://www.pythonchallenge.com/pc/def/banner.p"
content = urlopen(u).read()
data = pickle.loads(content)

print(data)
````

大略看了一下，list 中的項目有空格也有#，看起來很像是點矩陣式的資料

算了一下每個項目的數量加起來都是 95，應該就是每行有 95 個字元，來印印看

````python
for item in data:
    print(*[x[0] * x[1] for x in item], sep="")
````
